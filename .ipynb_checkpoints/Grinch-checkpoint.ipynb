```python
def reformatRules():    #s string turned into dictionary dict
    infile = open (r'grinchRules.txt' ,"r")
    s = infile.read()#reads the entire files
    infile.close()
    

    #import re
    newString = re.sub(r'([.?!]) (%X [a-zA-Z \']+ %Y) ([A-Z])',r'\1key:\2value:\3',s)#returns a string 


    patternObj = re.compile(r'key:%X ([a-zA-Z \']+) %Yvalue:')#creates a pattern object of the RE
    list = patternObj.split(newString)#returns a list
    #print(len(list))#odds == keys  #evens == unsplit values

    dict={}#dict[key] = value;

    for i in range(1, len(list), 2):
        newString = re.sub(r'([.|?]) ([A-Z|%])',r'\1value_\2',list[i+1])#returns a string 
        patternObj = re.compile(r'value_')#creates a pattern object of the RE    
        newList = patternObj.split(newString)#returns a list
        dict[list[i]] = newList
    return dict

def checkResponse(key,response,s):
    
    
    stringX = ""
    stringY = ""


    #create %X
    patternObj = re.compile(key)
    list = patternObj.split(s)  
    if list[0]:
        stringX = list[0]
        stringX = translateInput(stringX)
        if len(stringX) != 0:
            if stringX[0] != "I":
                stringX = stringX[0].lower() + stringX[1:-1]
            else: stringX = stringX[:-1]
   
    #create %Y 
    if len(list) == 2: 
        stringY = list[1]
        stringY = translateInput(stringY)
        stringY = stringY[:-1]
    
    response = re.sub('%X',stringX,response)
    response = re.sub(' %Y',stringY,response)
    
    response = response[0].upper() + response[1:]

    return response
    
    

def translateInput(s):    
    #replaces pronouns
    import re

    pronounDict = {}
    pronounDict['i'] = 'you'                # I - you
    pronounDict['me'] = 'you'               # me - you
    pronounDict['my'] = 'your'              # my - your
    pronounDict['mine'] = 'yours'           # mine - yours
    pronounDict['we'] = 'you'               # we - you
    pronounDict['us'] = 'you'               # us - you
    pronounDict['our'] = 'your'             # our - your
    pronounDict['ours'] = 'yours'           # ours - yours

    pronounDict['you'] = 'I'                # you - I
    pronounDict['your'] = 'my'              # your - my
    pronounDict['are you'] = 'am I'         # are you - am i

    # @end of sentence or beginning with commas
    pronounDict['yours.'] = 'mine'         # yours - mine
    pronounDict['you.'] = 'me'             # you - me
    pronounDict['well, '] = ''
    pronounDict['so, '] = ''

    s = s.lower()
    

    patternObj = re.compile(r'([a-zA-Z]+)')#creates a pattern object of the RE
    list = patternObj.split(s)#returns a list
    

    # @end of sentences or beginning with commas
    if len(list) > 1:
        if list[-2] == 'yours':
            list[-2] = 'yours.'

        if list[-2] == 'you':
            list[-2] = 'you.'
        
        if list[1] == 'well' and list[2] == ', ':
            list[1] = 'well, '
            list = list[0:1] + list[3:]   
        elif list[1] == 'so' and list[2] == ', ':
            list[1] = 'so, '
            list = list[0:1] + list[3:]

    temp = 0
    for i in list:
        if i in pronounDict:
            list[temp] = pronounDict[i]
        temp += 1


    
    s =''
    for i in list:        
        s = s + i
    
    if s=='':
        return ''
    else: 
        return s[0].upper() + s[1:]
    


import random
import re

dict = reformatRules()

#print(dict) debug dictionary

response = ''
s = ''
s = input('User: ')
while (s != 'quit'):
    s = s.strip()#remove extra whitespaces
    if s.find(';') != -1:#returns -1 if not found
        s = s[:s.find(';')]#keeps part before ;
    
    
    if s[0:2] != "I ":#if sentence starts with I
        s = s[0].lower() + s[1:]#replace I with i
        
    pattern = re.compile('\w')#create a patter of alphanumeric characters
    m = pattern.match(s[-1])#checks last sentence for no punctuation
    if m:
        s = s + '.' #replaces last of string with punctuation
  
    temp = []
    for key in dict.keys():
        pattern = re.compile('\W*' + key + '[ \W]') #creates a patter to search for all key words in for loop
        m = pattern.search(s)
        
        if m: #if keyword is present
            temp = dict[key]
            response = dict[key][random.randint(0,len(temp)-1)]
            response = checkResponse(key,response,s)
            break
        if key == 'default':#if keyword is not present
            #debugging here was: response = translateInput(s)
            temp = dict[key]
            response = dict[key][random.randint(0,len(temp)-1)]
            response = checkResponse(key,response,s)
            break
    
    s = input('Grinch: ' + response + '\nUser: ')
    
    
```
